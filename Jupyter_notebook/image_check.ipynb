### 1.사용할 이미지 데이터들을 확인하기위해서 간단하게 먼저 이미지를 바꿔준후 시각화를 진행하겠습니다.
1) 여기서 이미지를 BGR에서 RGB로 변환을 하게된것은. 옛날에 이미지를 저장할떄 BGR형태로 저장을 했다해서 진행을 했습니다.

```python
import os
import glob
import random
import math
import shutil #shutil 모듈은 파일을 원하는 경로에 복사해줌.

import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import cv2
%matplotlib inline
```

```python
path = "/Users/sunghoheo/last_project/AI_05_허성호"
black = glob.glob(path+"/black" + "/*")
picture = glob.glob(path+"/picture" + "/*")
home_alone = glob.glob("./home_alone" + "/*")
```

```python
def read_image(file):
    image_path = cv2.imread(file)
    return cv2.cvtColor(image_path, cv2.COLOR_BGR2RGB)
```

```python
image_arr = []
image_num = range(0,200)

for i in random.sample(image_num,3):
    image_arr.append(read_image(home_alone[i]))
    image_arr.append(read_image(black[i]))
    image_arr.append(read_image(picture[i]))
```

```python
print("총 시각화할 이미지 갯수:",len(image_arr))
```

```python
rows = 3
columns = 3

fig,axes = plt.subplots(nrows=rows, ncols=columns, figsize=(columns*3,rows*3))

#각 그림판에 이미지 출력하기
for num in range(1, rows*columns+1):
    
    fig.add_subplot(rows, columns, num)
    idx =  num-1
    
    plt.imshow(image_arr[idx],aspect='auto')
    plt.xlabel(f'{image_arr[idx].shape}', fontsize=12)

fig.tight_layout()

cols = ['home_alone','black','picture']

for folder_idx, ax in enumerate(axes[0]):
            ax.set_title(cols[folder_idx])
        
for idx,ax in enumerate(axes.flat):
               ax.set_xticks([])
               ax.set_yticks([])
               
```
